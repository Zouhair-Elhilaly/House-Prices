3- Data Processing
Ex 3.1

```python
missing_data[missing_data.Percentage > 50]
```

```python
df.drop( columns=missing_data[missing_data.Percentage > 50].index, inplace=True)
```

```python
print("size of columns removed as " ,len(missing_data[missing_data.Percentage > 50]))
print("new shape is : ", df.shape)
```

```python
for col in df:
    if df[col].isnull().sum() > 0:  
        if df[col].dtype == 'object':  
            df[col].fillna(df[col].mode()[0], inplace=True)
        else:  
            df[col].fillna(df[col].median(), inplace=True)
```

ex 3.3

```python
x  = df.drop(columns=['SalePrice'])
y = df['SalePrice']
x.shape, y.shape
```

Ex 3.4 encode categorical variables 

```python
def onehot_encode(df):
    cat_columns = df.select_dtypes(include='object').columns
    df_encoded = pd.get_dummies(df, columns=cat_columns, drop_first=True)
    return df_encoded
x_encoded = onehot_encode(x)
print("NEW SHAPE is : ",x_encoded.shape)
```

Ex  3.5

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x_encoded.select_dtypes(include=[np.number]))
```

# 4 Train - test split
Ex 4.1

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=42)
```

# 5 Linear Regression
Ex 5.1

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("Mean Squared Error:", mse)
print("R^2 Score:", r2)
```

# 8 Compare Models 
8.1

```python
models = {
    "Linear Regression": LinearRegression(),
    "Ridge": Ridge(alpha=1.0),
    "Lasso": Lasso(alpha=0.1)
}

results = []

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    r2 = r2_score(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    
    results.append({
        "Model": name,
        "RÂ²": r2,
        "MSE": mse
    })

metrics_df = pd.DataFrame(results)
print( "*"*80 ,"\n\n",metrics_df,"\n")
print("*"*80)
```

8.2

```python
sorted_df = metrics_df.sort_values(by='RÂ²', ascending=False)
print(sorted_df)
print("The best model is : ",sorted_df.Model.values[0])
```
